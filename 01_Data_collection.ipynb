```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stocks = ["MSFT","AAPL","AMZN","NVDA","GOOGL"]
```

```python
dfs=[]

for stock in stocks:
    df= yf.download(stock,start="2020-01-01", end="2025-12-31",
                    auto_adjust=False, progress= False)

    if isinstance(df.columns, pd.MultiIndex):
        df.columns= df.columns.get_level_values(0)


    df = df.reset_index()
    df["Ticker"]= stock

    dfs.append(df)

all_data=pd.concat(dfs, ignore_index=True)

all_data.head()




```

```python
all_data.info()
```

```python
all_data.isnull().sum()
```

```python
#Daily return percentage

all_data["Daily_Return"]= (all_data.groupby("Ticker")["Close"].pct_change()*100)
```

```python
The daily return percentage has a pattern of minor dip and minor rise. The fluctuations are not major differences hence it is less threatning.
```

```python
#Price change

all_data["Price_change"]=( all_data["Close"] - all_data["Open"] )
```

```python
So by observing the price change data, it is having minor fluctuations.
```

```python
# 20 Day Moving AVerage

all_data["20MA"]=( all_data.groupby("Ticker")["Close"].transform
                   (lambda x: x.rolling(20).mean())
                 )
```
